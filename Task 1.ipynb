```python
import matplotlib.pyplot as plt

# Data from the image
age_groups = ['0–20 Years', '21–64 Years', '65+ Years']
population_millions = [512, 807, 98]

# Create the bar chart
plt.figure(figsize=(9, 5))
plt.bar(age_groups, population_millions)

# Labels and title
plt.xlabel('Age Groups')
plt.ylabel('Population (in Millions)')
plt.title("India's Population Distribution by Age (2022)")

# Display values on top of bars
for i, value in enumerate(population_millions):
    plt.text(i, value + 10, f'{value} Mn', ha='center')

# Show the chart
plt.show()
```
